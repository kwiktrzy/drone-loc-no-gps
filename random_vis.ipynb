```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimb
import os
```

```python
path_left = "/workspace/datasets/train_tiles_one_to_one/Yunnan/patch__24_663482__102_361384__24_660321__102_364989__88d95323.jpg"
path_right = "/workspace/datasets/train_tiles_one_to_one/Yunnan/patch__24_66193124__102_3632073__2a0257eda23c492f81d89de0e08294b0_1.jpg"
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(12,6))

if os.path.exists(path_left):
    img1 = mpimb.imread(path_left)
    ax1.imshow(img1)
    ax1.set_title("Left")
    ax1.axis("off")
else: 
    ax1.text(0.5, 0.5, "No file", ha='center')

if os.path.exists(path_right):
    img2 = mpimb.imread(path_right)
    ax2.imshow(img2)
    ax2.set_title("Right")
    ax2.axis("off")
else: 
    ax2.text(0.5, 0.5, "No file", ha='center')

plt.tight_layout()
plt.show()
```
